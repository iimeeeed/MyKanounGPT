```python
import requests
import os

# Base URL template
base_url = 'https://www.joradp.dz/FTP/jo-arabe/{}/A{}{:03d}.pdf'

# Output base directory
output_base_dir = 'joradp'
os.makedirs(output_base_dir, exist_ok=True)

# Loop through years
for year in range(1964, 2025):
    # Create a subfolder for each year
    year_dir = os.path.join(output_base_dir, str(year))
    os.makedirs(year_dir, exist_ok=True)
    
    # Flag to stop downloading for this year
    found_pdf = True
    
    for paper_number in range(1, 365):  # Large number to cover potential range
        if not found_pdf:
            break
        
        # Generate the URL
        url = base_url.format(year, year, paper_number)
        
        # Send the request
        response = requests.get(url)
        
        # Check if the PDF exists
        if response.status_code == 200:
            # Save the PDF
            file_name = f'A{year}{paper_number:03d}.pdf'
            file_path = os.path.join(year_dir, file_name)
            with open(file_path, 'wb') as file:
                file.write(response.content)
            print(f'Downloaded: {file_name}')
        elif response.status_code == 404:
            print(f'Not found: {url}')
            found_pdf = False  # Stop processing further PDFs for this year
        else:
            print(f'Error {response.status_code} for: {url}')
```
