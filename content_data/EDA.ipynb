```python
import json

# Load the combined JSON data from the file
with open('combined_restructured_data.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

# Counters for different lengths of 'details'
count_four_details = 0
count_three_or_less_details = 0

# Iterate over each object in the data
for item in data:
    if 'details' in item:
        details_length = len(item['details'])
        if details_length == 4:
            count_four_details += 1
        elif details_length <= 3:
            count_three_or_less_details += 1
            print(item['details'])

# Print the results
print(f"Number of objects with 4 items in 'details': {count_four_details}")
print(f"Number of objects with 3 or fewer items in 'details': {count_three_or_less_details}")
```
