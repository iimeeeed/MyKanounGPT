# **Combined JSON Cleaning**


```python
import json
import glob
import os
import re

# Define the directory containing the JSON files
directory = "content_json"  # Update this path to your directory

# Define the pattern to match your JSON files
pattern = os.path.join(directory, "page_dval_*.json")


# Custom function to extract the numeric part (including negative) from the filename
def extract_number(filename):
    match = re.search(r"(-?\d+)", filename)  # Handle negative numbers
    if match:
        return int(match.group(1))
    return 0


# Get a sorted list of all files matching the pattern
file_list = sorted(glob.glob(pattern), key=extract_number)

# List to store all the data
combined_data = []

# Debug: Print the files being processed
print("Files to process in order:")
for file_name in file_list:
    print(f" - {file_name}")

# Read each JSON file and append its data to the combined_data list
for file_name in file_list:
    try:
        with open(file_name, "r", encoding="utf-8") as file:
            data = json.load(file)
            combined_data.extend(data)
            print(f"Loaded {len(data)} items from {file_name}")
    except Exception as e:
        print(f"Error processing {file_name}: {e}")

# Write the combined data to a new JSON file
output_file = "combined_file.json"
try:
    with open(output_file, "w", encoding="utf-8") as file:
        json.dump(combined_data, file, ensure_ascii=False, indent=4)
    print(f"Combined data written to {output_file}")
except Exception as e:
    print(f"Error writing to {output_file}: {e}")
```

```python
import re

combined_json_final = deepcopy(combined_json_transformed_cleaned)


def extract_year_number(string):

    # Regex to extract number and year
    match = re.search(r"عدد (\d+).* (\d{4})", string)

    if match:
        number = match.group(1)
        year = match.group(2)
        return pd.Series([year, number])
    return pd.Series([None, None])


combined_json_final[["year", "number"]] = combined_json_final["gazette"].apply(
    extract_year_number
)
combined_json_final
```

```python
import numpy as np
from copy import deepcopy

# Make a deep copy of the original DataFrame
combined_json_flattened = deepcopy(combined_json_final)

# Initialize the children column as an empty list for every row
combined_json_flattened["modified_by"] =None
combined_json_flattened["cancelled_by"] =None
combined_json_flattened["approved_by"] =None
combined_json_flattened["rectified_in"] = None
combined_json_flattened["implemented_by"] =None
first_parent = None

# List to store indices of rows to drop after nesting
to_drop = []
```

```python
# Nest level 2 inside level 1
for i in range(0, len(combined_json_flattened)):
    if combined_json_flattened.loc[i, "level"] == 1:
        # When encountering a level 1, mark it as the first parent
        first_parent = i
        combined_json_flattened.at[first_parent, "modified_by"] = []
        combined_json_flattened.at[first_parent, "approved_by"] = []
        combined_json_flattened.at[first_parent, "cancelled_by"] = []
        combined_json_flattened.at[first_parent, "rectified_in"] = []
        combined_json_flattened.at[first_parent, "implemented_by"] = []

    elif combined_json_flattened.loc[i, "level"] == 2:
        # Add level 2 entries (which already contain their children) as children of level 1
        child_row = combined_json_flattened.loc[i].to_dict()
        # child_row.pop("level", None)
        # child_row.pop("action", None)
        # child_row.pop("modified_by", None)
        # child_row.pop("approved_by", None)
        # child_row.pop("cancelled_by", None)
        # child_row.pop("rectified_in", None)
        # child_row.pop("implemented_by", None)

        if combined_json_flattened.loc[i, "action"] == "معدل بـ":
            combined_json_flattened.at[first_parent, "modified_by"].append(child_row)
            if i < 20:
                print(f"Added {i} to {first_parent} at modified")
        elif combined_json_flattened.loc[i, "action"] == "نص تطبيقي":
            combined_json_flattened.at[first_parent, "implemented_by"].append(child_row)
            if i < 20:
                print(f"Added {i} to {first_parent} at implemented_by")
        elif combined_json_flattened.loc[i, "action"] == "ملغى بـ":
            combined_json_flattened.at[first_parent, "cancelled_by"].append(child_row)
            if i < 20:
                print(f"Added {i} to {first_parent} at cancelled_by")
        elif combined_json_flattened.loc[i, "action"] == "استدراك في":
            combined_json_flattened.at[first_parent, "rectified_in"].append(child_row)
            if i < 20:
                print(f"Added {i} to {first_parent} at rectified_in")
        else:
            combined_json_flattened.at[first_parent, "approved_by"].append(child_row)
            if i < 20:
                print(f"Added {i} to {first_parent} at approved_by")

        to_drop.append(i)

# Drop level 2 and 3 rows (children) from the original DataFrame after nesting
combined_json_flattened.drop(index=to_drop, inplace=True)

# Reset index to maintain sequential order after dropping rows
combined_json_flattened.reset_index(drop=True, inplace=True)

combined_json_flattened.drop(columns=["level", "action"], inplace=True)

# Display the first few rows to verify
combined_json_flattened.head(10)
```

```python
import json
import pandas as pd

final_json = []

# Loop through the flattened JSON array
for i in range(len(combined_json_flattened)):
    current_obj = combined_json_flattened.iloc[i]

    # If the current object is a pandas Series, convert it to a dictionary
    if isinstance(current_obj, pd.Series):
        current_obj = current_obj.to_dict()

    year = current_obj.get('year')
    number = current_obj.get('number')

    # Check if there is already an entry for this year and number
    existing_entry = next((item for item in final_json if item['year'] == year and item['number'] == number), None)

    if existing_entry:
        # If entry exists, append to the content
        existing_entry['content'].append(current_obj)
    else:
        # If entry doesn't exist, create a new entry
        final_json.append({
            "year": year,
            "number": number,
            "content": [current_obj]
        })

# Write the final_json object to a file
with open('final_json_output.json', 'w', encoding='utf-8') as file:
    json.dump(final_json, file, ensure_ascii=False, indent=4)

print("JSON data successfully written to final_json_output.json")
```
